```python
import yfinance as yf
import pandas as pd

ticker = yf.Ticker("QQQ")

# get all stock info
ticker.info

# get historical market data
hist = ticker.history(period="max")

# Calculate 5-day Simple Moving Average (SMA)
hist['5D SMA'] = hist['Close'].rolling(window=5).mean()

# Calculate 13-day Simple Moving Average (SMA)
hist['13D SMA'] = hist['Close'].rolling(window=13).mean()

# Calculate Area btw 5-day Simple Moving Average and 13-day Simple Moving Average
hist['Area btw'] = hist['5D SMA'] - hist['13D SMA']

# Calculate 5-day Average Volume
hist['5DA Volume'] = hist['Volume'].rolling(window=5).mean()

# Calculate 13-day Average Volume
hist['13DA Volume'] = hist['Volume'].rolling(window=13).mean()

# Calculate 25-day Average Volume
hist['25DA Volume'] = hist['Volume'].rolling(window=25).mean()

# Caclulate 5-day Enter Signal
hist['5DA Enter Signal'] = ((hist['Volume'] > hist['Volume'].shift(1)) & (hist['5DA Volume'] > hist['5DA Volume'].shift(1))).astype(int)

# Caclulate 13-day Enter Signal
hist['13DA Enter Signal'] = ((hist['Volume'] > hist['Volume'].shift(1)) & (hist['13DA Volume'] > hist['13DA Volume'].shift(1))).astype(int)

# Caclulate 25-day Enter Signal
hist['25DA Enter Signal'] = ((hist['Volume'] > hist['Volume'].shift(1)) & (hist['25DA Volume'] > hist['25DA Volume'].shift(1))).astype(int)

#Calculate Total Enter Signal
hist['Total Enter Signal'] = hist['5DA Enter Signal'] + hist['13DA Enter Signal'] + hist['25DA Enter Signal']

hist

# show meta information about the history (requires history() to be called first)
#qqq.history_metadata
```
